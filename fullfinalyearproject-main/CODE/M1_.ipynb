```python
from  sklearn.preprocessing  import LabelEncoder
le = LabelEncoder()

var = ['test_date', 'cough', 'fever', 'sore_throat', 'shortness_of_breath',
       'head_ache', 'corona_result', 'age_60_and_above', 'gender',
       'test_indication']
for x in var:
    data[x] = le.fit_transform(data[x])
```
